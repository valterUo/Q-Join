```python
tables = [0,1,2,3]
joins = [(0, 1), (1, 2), (2, 3)]
selectivity = {(0, 1): 1, (1, 2): 1, (2, 3): 1}
cardinality = {0: 100, 1: 100, 2: 100, 3: 100}
levels = range(3)
binpoly = {}

for level in levels:
    binpoly[level] = {}
    for join in joins:  
        if level == 0:
            binpoly[level][join] = (selectivity[join]*cardinality[join[0]] + selectivity[join]*cardinality[join[1]], [(level, join)])
        else:
            for table in join:
                prev_joins = list(binpoly[level - 1].keys())
                for prev_join in prev_joins:
                    if table in prev_join and prev_join != join:
                        prev_coeff = binpoly[level - 1][prev_join][0]
                        prev_variables = list(binpoly[level - 1][prev_join][1])
                        
        
            
            for prev_join in prev_joins:
                if join == prev_join:
                    continue
                #print(join, prev_join)
                prev_coeff = binpoly[level - 1][prev_join][0]
                variables = list(binpoly[level - 1][prev_join][1])
                
                coeff = selectivity[join]*cardinality[prev_join[0]]*prev_coeff + selectivity[join]*cardinality[prev_join[1]]*prev_coeff #+ prev_coeff
                
                variables.append((level, join))
                
                #binpoly[level][join] = (coeff, variables)
                if join in binpoly[level]:
                    coeffs = binpoly[level][join][0]
                    coeffs.append(coeff)
                    old_variables = binpoly[level][join][1]
                    old_variables.append(variables)
                    binpoly[level][join] = (coeffs, old_variables)
                else:
                    print(coeff, variables)
                    binpoly[level][join] = ([coeff], [variables])
```

```python
for key in binpoly:
    print()
    print("Level: ", key)
    for key2 in binpoly[key]:
        print("Join: ", key2)
        print("Coefficient: ", binpoly[key][key2][0])
        print("Variables: ", binpoly[key][key2][1])
```
