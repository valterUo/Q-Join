```python
import numpy as np
import matplotlib.pyplot as plt

# Defining the range of relations
relations = np.arange(3, 61)

# Defining the functions for variables
def clique(relations):
    return (relations - 1) * (relations * (relations - 1)) / 2 + relations * np.log2(relations - 1)

def chain(relations):
    return (relations - 1) ** 2

def star(relations):
    return (relations - 1) ** 2

def tree(relations):
    return (relations - 1) ** 2

def cycle(relations):
    return (relations - 1) * relations

def baseline(relations):
    return 2 * relations * (relations - 1) + relations + 3 * (relations - 1) * ((relations - 1) - 1)

def novel(relations):
    return (relations + (relations - 1)) * ((relations - 1) - 1)

def baseline_cycle(relations):
    return 2 * relations * (relations) + relations + 3 * (relations) * ((relations) - 1)

def novel_cycle(relations):
    return (relations + (relations)) * ((relations) - 1)

# Handling the invalid values (log(0) is undefined for clique)
def safe_clique(relations):
    return np.where(relations > 1, clique(relations), 0)

def nitin_groppe(relations):
    return 2**relations - 1

# Plotting the functions
# Define column width in inches (for example, 240pt column width)
column_width_pt = 240  # Adjust this value to your LaTeX column width
inches_per_pt = 1 / 72.27  # Conversion factor: points to inches

# Column width in inches
column_width_in = 8 #column_width_pt * inches_per_pt

# Adjust aspect ratio (e.g., golden ratio)
aspect_ratio = 0.618  # You can change this value based on your needs
plt.figure(figsize=(column_width_in, column_width_in * aspect_ratio))

# Setting background color to a lighter gray
plt.gca().set_facecolor('#eaeaea')  # Lighter gray

# Plotting with dots and scaling by dividing by 1000 to show values in thousands
#plt.plot(relations, safe_clique(relations) / 1000, 'o', label='Clique', color='blue')
#plt.plot(relations, chain(relations) / 1000, 'o', label='Chain', color='green')
#plt.plot(relations, star(relations) / 1000, 'o', label='Star', color='red')
#plt.plot(relations, tree(relations) / 1000, 'o', label='Tree', color='orange')
plt.plot(relations, cycle(relations) / 1000, '-o', label='Q-Join - Cycle', color='purple', markersize=4)
plt.plot(relations, baseline_cycle(relations) / 1000, '-o', label='Schnberger et al. (2023) - Cycle', color='brown', markersize=4)
plt.plot(relations, novel_cycle(relations) / 1000, '-o', label='Schnberger et al. (2024) - Cycle', color='cyan', markersize=4)
#plt.plot(relations, nitin_groppe(relations) / 1000, '-o', label='Nitin et al. (2024)', color='black', markersize=4)

# Adding labels and legend
plt.xlabel('# Relations', fontdict={'fontsize': 14})
plt.ylabel('# Variables (in thousands)', fontdict={'fontsize': 14})
#plt.title('Relations-Variables Plane for Different Functions (scaled in thousands)')
plt.legend(fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.grid(True, color='white')

# Customizing grid and ticks for visibility on lighter gray background
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.xlim(3, 60)

# Display the plot
plt.show()
```
